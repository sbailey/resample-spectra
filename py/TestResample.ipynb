```python
import numpy as N
from scipy.sparse.construct import spdiags

def resolution_from_icov(iCov):
    """
    Given inverse covariance matrix iCov, returns (R, ivar) such that
        Cov = (iCov)^-1
        diagCov = R Cov R.T = diag(1/ivar)

    R is the "resolution matrix" which transforms a vector with inverse
    covariance iCov into a basis with diagonal errors given by inverse
    variances ivar.  See Bolton & Schlegel 2010 (B&S) Equations 10-15
        
    Adapted from A. Bolton, Utah
    """
    #- Solve iCov = QQ
    w, V = N.linalg.eigh(iCov)
    Wx = spdiags(N.sqrt(w), 0, len(w), len(w))
    Q = V.dot( Wx.dot(V.T) )

    #- Create resolution matrix R
    norm = N.sum(Q, axis=1)                       #- B&S Eqn 11
    R = N.outer(1.0/norm, N.ones(norm.size)) * Q  #- B&S Eqn 12
    return R, norm**2
```

```python
from scipy import sparse
import scipy.sparse.linalg


#- Projection matrix for linear interpolation
#- y = A yfit sampled at x and xfit
def getF(w1, w2):
    """
    Get matrix F such that w1 = F w2 by linear interpolation
    
    This is used to transform data on wavelength grid (w2) to grid (w1)
    """
    n1 = len(w1)
    n2 = len(w2)
    F = sparse.lil_matrix((n1, n2))
    
    #- At what indices would w1 fit into w2?
    jj = N.searchsorted(w2, w1)
    
    for i, j in enumerate(jj):
        if j == 0 or j == n2:
            continue
            
        dx = (w1[i] - w2[j-1]) / (w2[j] - w2[j-1])
        F[i, j-1] = (1-dx)
        F[i, j] = dx
        
    return F.tocsr()
```

# Example 1: Resample a single spectrum

```python
#- Generate some data
import numpy as N
N.random.seed(0)    #- make it randomly reproducible

xmax = 20
x = N.linspace(0.0, xmax, 30)
err = N.ones(len(x))
y = N.random.normal(scale=err)
```

```python
#- Solve for model on a different sampling xx
#- Ignores errors for the moment
xx = N.linspace(-0.0, xmax, 20)
F = getF(x, xx).toarray()
W = N.diag(1.0/err**2)
iCov = F.T.dot(W.dot(F))
yy = N.linalg.solve(iCov, F.T.dot(W.dot(y)))
R, ivar = resolution_from_icov(iCov)
```

```python
#- Plots
errorbar(x, y, err, fmt='bs', capsize=0, alpha=0.5, label='Input data')
# plot(xx, R.dot(yy), 'g^-', label='Diagonalized model')
errorbar(xx, R.dot(yy), 1/N.sqrt(ivar), fmt='g^-', label='Diagonalized model')
plot(xx, yy, 'rs-', label='Covariant model')
ylim(-5, 9)
xlim(-1, xmax+1)
axhline(0.0, color='k', alpha=0.1)
legend()
savefig('single_spec_1.pdf')
```

```python
#- Show non-diagonalized covariance matrix
set_cmap('RdBu_r')
subplot(131)  
imshow(N.linalg.inv(iCov), origin='upper', vmin=-3, vmax=3)
xlabel('Covariance')
subplot(132)
imshow(iCov, origin='upper')
xlabel('Inverse Covariance')
subplot(133)
imshow(R, origin='upper')
xlabel('Resolution Matrix')
```

# Example 2: Coadd unaligned spectra

```python
#- Generate two spectra with different x sampling
N.random.seed(0)
def model(x):
    return N.sin(x)

x1 = N.linspace(0, 2*N.pi, 30)
x2 = N.linspace(0, 2*N.pi, 31)
e1 = 0.1*N.ones(len(x1))
e2 = 0.1*N.ones(len(x2))
y1 = model(x1) + N.random.normal(scale=e1)
y2 = model(x2) + N.random.normal(scale=e2)

errorbar(x1, y1, e1, fmt='bs', capsize=0, alpha=0.5)
errorbar(x2, y2, e2, fmt='cs', capsize=0, alpha=0.5)
xlim(-0.1, 2*N.pi+0.1)
ylim(-1.5, 1.5)
```

```python
#- Coadd the spectra
xc = N.linspace(-0.1, 2*N.pi+0.1, 30)
yt = model(xc)

y = N.concatenate([y1, y2])
err = N.concatenate([e1, e2])
W = N.diag(1.0/err**2)
F1 = getF(x1, xc).toarray()
F2 = getF(x2, xc).toarray()
F = N.vstack([F1, F2])

iCov = F.T.dot(W.dot(F))
yc = N.linalg.solve(iCov, F.T.dot(W.dot(y)))
R, ivar = resolution_from_icov(iCov)
ec = N.sqrt(1/ivar)

errorbar(x1, y1-model(x1), e1, fmt='bs', capsize=0, alpha=0.2, label='Spectrum 1')
errorbar(x2, y2-model(x2), e2, fmt='cs', capsize=0, alpha=0.2, label='Spectrum 2')
plot(xc, yc-model(xc), 'rs-', label='Covariant model')
plot(xc, R.dot(yc-model(xc)), 'gs-', label='Diagonalized model')
title('Residuals from input spectrum')
xlim(-0.1, 2*N.pi+0.1)
# ylim(-0.3, 0.6)
legend()
axhline(0.0, color='k', alpha=0.1)
```

```python
Cov = N.linalg.inv(iCov)
subplot(131)  
imshow(Cov, origin='upper')
xlabel('Covariance')
subplot(132)
imshow(iCov, origin='upper')
xlabel('Inverse Covariance')
subplot(133)
imshow(R, origin='upper')
xlabel('Resolution Matrix')
```

### Example 2a: Test covariance with larger sample

```python
#- Generate two spectra=0 with different x sampling
N.random.seed(0)
n = 300.0
x1 = N.arange(n)
x2 = N.arange(n)+0.01
e1 = 0.1*N.ones(len(x1))
e2 = 0.1*N.ones(len(x2))
# y1 = N.random.normal(scale=e1)
# y2 = N.random.normal(scale=e2)
x0 = n/2.
sigma = n/30.
y1 = exp(-(x1-x0)**2/(2*sigma**2)) + N.random.normal(scale=e1)
y2 = exp(-(x1-x0)**2/(2*sigma**2)) + N.random.normal(scale=e2)

plot(x1, y1, 'b,')
plot(x2, y2, 'r+')

ylim(-0.5,1.5)
axhline(0.0, color='k', alpha=0.1)
```

```python
#- Coadd on different x sampling
xc = N.linspace(1, n-1, 295)
yt = model(xc)

y = N.concatenate([y1, y2])
err = N.concatenate([e1, e2])
W = N.diag(1.0/err**2)
F1 = getF(x1, xc).toarray()
F2 = getF(x2, xc).toarray()
F = N.vstack([F1, F2])

iCov = F.T.dot(W.dot(F))
yc = N.linalg.solve(iCov, F.T.dot(W.dot(y)))
R, ivar = resolution_from_icov(iCov)
ec = N.sqrt(1/ivar)
```

```python
subplot(211)
plot(x1, y1, 'bx')
plot(x2, y2, 'c+')
plot(xc, yc, 'r-')
plot(xc, R.dot(yc), 'g-')
axhline(0.0, color='k', alpha=0.1)
ylim(-0.5,1.5)
axvspan(175, 200, color='k', alpha=0.1)

#- Highlight covariant ringing
subplot(212)
plot(x1, y1, 'bx')
plot(x2, y2, 'c+')
plot(xc, yc, 'r.-')
# plot(xc, R.dot(yc), 'g-')
errorbar(xc, R.dot(yc), ec, fmt='g.-', capsize=0)
axhline(0.0, color='k', alpha=0.1)
ylim(-0.75,0.75)
xlim(175, 200)

savefig('coadd_spectra.pdf')
```

# Example 3: Resample covariant spectra

```python
#- Generate inverse covariance for neighboring spectra
nspec = 3
nx = 3*(nspec+1)
A = N.zeros((nspec, nx))
x = N.arange(nx, dtype=float)
for i in range(nspec):
    #- A real PSF would integrate the gaussian rather than resample, but ok for demo
    A[i] = N.exp( -(x-nspec*(i+1))**2 / 2 )

Cinv0 = A.dot(A.T)

nflux = 20
Cinv = N.zeros( (nspec*nflux, nspec*nflux) )
for i in range(nspec):
    for j in range(nspec):
        for k in range(nflux):
            Cinv[i*nflux+k,j*nflux+k] = Cinv0[i,j]
            
imshow(Cinv)
C = N.linalg.inv(Cinv)
```

```python
#- Generate some uncorrelated spectra and then rotate them into the correlated basis
V, S, VT = N.linalg.svd(C)
f = N.zeros((1000, nspec*nflux))
for i in range(f.shape[0]):
    f[i] = VT.dot(N.random.normal(size=nspec*nflux))

subplot(121); imshow(f.T.dot(f))
subplot(122); imshow(C)
```
