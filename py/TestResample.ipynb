```python
import numpy as N
from scipy.sparse.construct import spdiags

def resolution_from_icov(iCov):
    """
    Given inverse covariance matrix iCov, returns (R, ivar) such that
        Cov = (iCov)^-1
        diagCov = R Cov R.T = diag(1/ivar)

    R is the "resolution matrix" which transforms a vector with inverse
    covariance iCov into a basis with diagonal errors given by inverse
    variances ivar.  See Bolton & Schlegel 2010 (B&S) Equations 10-15
        
    Adapted from A. Bolton, Utah
    """
    #- Solve iCov = QQ
    w, V = N.linalg.eigh(iCov)
    Wx = spdiags(N.sqrt(w), 0, len(w), len(w))
    Q = V.dot( Wx.dot(V.T) )

    #- Create resolution matrix R
    norm = N.sum(Q, axis=1)                       #- B&S Eqn 11
    #- Equivalent but faster than:  R = N.diag(1.0/norm).dot(Q)
    R = N.outer(1.0/norm, N.ones(norm.size)) * Q  #- B&S Eqn 12
    return R, norm**2
```

```python
from scipy import sparse
import scipy.sparse.linalg


#- Projection matrix for linear interpolation
#- y = A yfit sampled at x and xfit
def getF(w1, w2):
    """
    Get matrix F such that w1 = F w2 by linear interpolation
    
    This is used to transform data on wavelength grid (w2) to grid (w1)
    """
    n1 = len(w1)
    n2 = len(w2)
    F = sparse.lil_matrix((n1, n2))
    
    #- At what indices would w1 fit into w2?
    jj = N.searchsorted(w2, w1)
    
    for i, j in enumerate(jj):
        if j == 0 or j == n2:
            continue
            
        dx = (w1[i] - w2[j-1]) / (w2[j] - w2[j-1])
        F[i, j-1] = (1-dx)
        F[i, j] = dx
        
    return F.tocsr()
```

# Example 1: Resample a single spectrum

```python
#- Generate some data
import numpy as N
N.random.seed(0)    #- make it randomly reproducible

xmax = 20
x = N.linspace(0.0, xmax, 30)
err = N.ones(len(x))
y = N.random.normal(scale=err)
```

```python
#- Solve for model on a different sampling xx
#- Ignores errors for the moment
xx = N.linspace(-0.0, xmax, 20)
F = getF(x, xx).toarray()
W = N.diag(1.0/err**2)
iCov = F.T.dot(W.dot(F))
yy = N.linalg.solve(iCov, F.T.dot(W.dot(y)))
R, ivar = resolution_from_icov(iCov)
```

```python
#- Plots
errorbar(x, y, err, fmt='bs', capsize=0, alpha=0.5, label='Input data')
# plot(xx, R.dot(yy), 'g^-', label='Diagonalized model')
errorbar(xx, R.dot(yy), 1/N.sqrt(ivar), fmt='g^-', label='Diagonalized model')
plot(xx, yy, 'rs-', label='Covariant model')
ylim(-5, 9)
xlim(-1, xmax+1)
axhline(0.0, color='k', alpha=0.1)
legend()
savefig('single_spec_1.pdf')
```

```python
#- Show non-diagonalized covariance matrix
set_cmap('RdBu_r')
subplot(131)  
imshow(N.linalg.inv(iCov), origin='upper', vmin=-3, vmax=3)
xlabel('Covariance')
subplot(132)
imshow(iCov, origin='upper')
xlabel('Inverse Covariance')
subplot(133)
imshow(R, origin='upper')
xlabel('Resolution Matrix')
```

# Example 2: Coadd unaligned spectra

```python
#- Generate two spectra with different x sampling
N.random.seed(0)
def model(x):
    return N.sin(x)

x1 = N.linspace(0, 2*N.pi, 30)
x2 = N.linspace(0, 2*N.pi, 31)
e1 = 0.1*N.ones(len(x1))
e2 = 0.1*N.ones(len(x2))
y1 = model(x1) + N.random.normal(scale=e1)
y2 = model(x2) + N.random.normal(scale=e2)

errorbar(x1, y1, e1, fmt='bs', capsize=0, alpha=0.5)
errorbar(x2, y2, e2, fmt='cs', capsize=0, alpha=0.5)
xlim(-0.1, 2*N.pi+0.1)
ylim(-1.5, 1.5)
```

```python
#- Coadd the spectra
xc = N.linspace(-0.1, 2*N.pi+0.1, 30)
yt = model(xc)

y = N.concatenate([y1, y2])
err = N.concatenate([e1, e2])
W = N.diag(1.0/err**2)
F1 = getF(x1, xc).toarray()
F2 = getF(x2, xc).toarray()
F = N.vstack([F1, F2])

iCov = F.T.dot(W.dot(F))
yc = N.linalg.solve(iCov, F.T.dot(W.dot(y)))
R, ivar = resolution_from_icov(iCov)
ec = N.sqrt(1/ivar)

errorbar(x1, y1-model(x1), e1, fmt='bs', capsize=0, alpha=0.2, label='Spectrum 1')
errorbar(x2, y2-model(x2), e2, fmt='cs', capsize=0, alpha=0.2, label='Spectrum 2')
plot(xc, yc-model(xc), 'rs-', label='Covariant model')
plot(xc, R.dot(yc-model(xc)), 'gs-', label='Diagonalized model')
title('Residuals from input spectrum')
xlim(-0.1, 2*N.pi+0.1)
# ylim(-0.3, 0.6)
legend()
axhline(0.0, color='k', alpha=0.1)
```

```python
Cov = N.linalg.inv(iCov)
subplot(131)  
imshow(Cov, origin='upper')
xlabel('Covariance')
subplot(132)
imshow(iCov, origin='upper')
xlabel('Inverse Covariance')
subplot(133)
imshow(R, origin='upper')
xlabel('Resolution Matrix')
```

```python
#- Generate two spectra=0 with different x sampling
N.random.seed(0)
n = 300.0
x1 = N.arange(n)
x2 = N.arange(n)+0.01
e1 = 0.1*N.ones(len(x1))
e2 = 0.1*N.ones(len(x2))
# y1 = N.random.normal(scale=e1)
# y2 = N.random.normal(scale=e2)
x0 = n/2.
sigma = n/30.
y1 = exp(-(x1-x0)**2/(2*sigma**2)) + N.random.normal(scale=e1)
y2 = exp(-(x1-x0)**2/(2*sigma**2)) + N.random.normal(scale=e2)

plot(x1, y1, 'b,')
plot(x2, y2, 'r+')

ylim(-0.5,1.5)
axhline(0.0, color='k', alpha=0.1)
```

```python
#- Coadd on different x sampling
xc = N.linspace(1, n-1, 295)
yt = model(xc)

y = N.concatenate([y1, y2])
err = N.concatenate([e1, e2])
W = N.diag(1.0/err**2)
F1 = getF(x1, xc).toarray()
F2 = getF(x2, xc).toarray()
F = N.vstack([F1, F2])

iCov = F.T.dot(W.dot(F))
yc = N.linalg.solve(iCov, F.T.dot(W.dot(y)))
R, ivar = resolution_from_icov(iCov)
ec = N.sqrt(1/ivar)
```

# Example 3: Resample covariant spectra

```python
#- Let's mimic a row-by-row extraction where
#- neighboring spectra are offset in wavelength
nspec = 4
nflux = 20

dx = 4     #- Separation between spectra
xsig = 1.5 #- PSF cross dispersion sigma
nx = dx*(nspec+1)

#- Single row projection matrix pix = A flux
A = N.zeros((nx, nspec))
x = N.arange(nx, dtype=float)
for i in range(nspec):
    x0 = dx*(i+1)
    A[:,i] = N.exp(-(x-x0)**2 / (2*xsig**2))
    A[:,i] /= N.sum(A[:,i])
```

```python
#- Create spectra
y = N.arange(nflux, dtype=float)
flux = N.zeros((nspec, nflux))
w = N.zeros((nspec, nflux))
wavemin = 8000.0
for i in range(nspec):
    offset = 0.6*i
    w0 = nflux/2 + offset
    w[i] = wavemin+y-offset
    flux[i] = 5 + 5*N.exp(-(y-w0)**2 / (2*2.0**2))

#- Project onto image
img = N.zeros((nflux, nx))
for row in range(nflux):
    img[row] = A.dot(flux[:,row])
    
#- Add noise to image
N.random.seed(0)
imgerr = 0.1*N.ones(img.shape)
noisyimg = img + N.random.normal(scale=imgerr)
                    
#- Note the offset in the peak position
imshow(noisyimg)
savefig('multispec_img.pdf')
```

```python
#- Extract flux from noisy image
xflux = N.zeros((nspec, nflux))
xCinv = N.zeros((nspec*nflux, nspec*nflux))
for row in range(nflux):
    Ninv = N.diag(1.0/imgerr[row]**2)
    y = A.T.dot(Ninv.dot(noisyimg[row]))
    Cinv = A.T.dot(Ninv.dot(A))
    xflux[:, row] = N.linalg.solve(Cinv, y)
    for i in range(Cinv.shape[0]):
        for j in range(Cinv.shape[1]):
            xCinv[i*nflux+row, j*nflux+row] = Cinv[i,j]
            
for i in range(nspec):
    # plot(w[i], flux[i])
    plot(w[i], xflux[i], drawstyle='steps-mid')
    
xlabel('Wavelength')
ylabel('Extracted flux')
savefig('multispec_extract.pdf')
```

```python
#- Now resample them to a different grid
# xp = N.arange(1.0, nflux-1.0, 1.2)
xp = N.arange(wavemin+1, wavemin+nflux-1, 1.2)
np = len(xp)
B = N.zeros([nspec*nflux, nspec*np])
for i in range(nspec):
    B[nflux*i:nflux*(i+1), np*i:np*(i+1)] = getF(w[i], xp).toarray()

#- y = B yp, with y having inverse covariance Cinv

biCov = B.T.dot( xCinv.dot(B) )
y = xflux.ravel()

yp = N.linalg.solve(biCov, B.T.dot(xCinv.dot(y))).reshape(nspec, np)

for i in range(nspec):
    yoffset = 2*i
    plot(w[i], xflux[i]+yoffset, drawstyle='steps-mid')
    plot(xp, yp[i]+yoffset, 'k.-')
    
ylim(0,18)
xlabel('Wavelength')
ylabel('Flux + offset')
title('Resampled spectra')
savefig('multispec_resample.pdf')
```

```python
#- Diagonalize
R, ivar = resolution_from_icov(biCov)
ypr = R.dot(yp.ravel()).reshape(nspec, np)
err = 1.0/N.sqrt(ivar.reshape(nspec, np))

for i in range(nspec):
    # plot(w[i], flux[i])
    yoffset = 2*i
    plot(w[i], xflux[i]+yoffset, drawstyle='steps-mid')
    # plot(xp, ypr[i]+yoffset, '.-')
    errorbar(xp, ypr[i]+yoffset, err[i], fmt='k.-')

ylim(0,18)
xlabel('Wavelength')
ylabel('Flux + offset')
title('Diagonalized resampled spectra')
savefig('multispec_diag.pdf')
```
