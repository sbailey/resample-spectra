```python
import numpy as N
from scipy.sparse.construct import spdiags

def resolution_from_icov(iCov):
    """
    Given inverse covariance matrix iCov, returns (R, ivar) such that
        Cov = (iCov)^-1
        diagCov = R Cov R.T = diag(1/ivar)

    R is the "resolution matrix" which transforms a vector with inverse
    covariance iCov into a basis with diagonal errors given by inverse
    variances ivar.  See Bolton & Schlegel 2010 (B&S) Equations 10-15
        
    Adapted from A. Bolton, Utah
    """
    #- Solve iCov = QQ
    w, V = N.linalg.eigh(iCov)
    Wx = spdiags(N.sqrt(w), 0, len(w), len(w))
    Q = V.dot( Wx.dot(V.T) )

    #- Create resolution matrix R
    norm = N.sum(Q, axis=1)                       #- B&S Eqn 11
    R = N.outer(1.0/norm, N.ones(norm.size)) * Q  #- B&S Eqn 12
    return R, norm**2
```

```python
from scipy import sparse
import scipy.sparse.linalg


#- Projection matrix for linear interpolation
#- y = A yfit sampled at x and xfit
def getF(w1, w2):
    """
    Get matrix F such that w1 = F w2 by linear interpolation
    
    This is used to transform data on wavelength grid (w2) to grid (w1)
    """
    n1 = len(w1)
    n2 = len(w2)
    F = sparse.lil_matrix((n1, n2))
    
    #- At what indices would w1 fit into w2?
    jj = N.searchsorted(w2, w1)
    
    for i, j in enumerate(jj):
        if j == 0 or j == n2:
            continue
            
        dx = (w1[i] - w2[j-1]) / (w2[j] - w2[j-1])
        F[i, j-1] = (1-dx)
        F[i, j] = dx
        
    return F.tocsr()
```

# Example 1: Resample a single spectrum

```python
#- Generate some data
import numpy as N
N.random.seed(0)    #- make it randomly reproducible

xmax = 20
x = N.linspace(0.0, xmax, 30)
err = N.ones(len(x))
y = N.random.normal(scale=err)
```

```python
#- Solve for model on a different sampling xx
#- Ignores errors for the moment
xx = N.linspace(-0.0, xmax, 20)
F = getF(x, xx).toarray()
W = N.diag(1.0/err**2)
iCov = F.T.dot(W.dot(F))
yy = N.linalg.solve(iCov, F.T.dot(W.dot(y)))
R, ivar = resolution_from_icov(iCov)
```

```python
#- Plots
errorbar(x, y, err, fmt='bs', capsize=0, alpha=0.5, label='Input data')
# plot(xx, R.dot(yy), 'g^-', label='Diagonalized model')
errorbar(xx, R.dot(yy), 1/N.sqrt(ivar), fmt='g^-', label='Diagonalized model')
plot(xx, yy, 'rs-', label='Covariant model')
ylim(-5, 9)
xlim(-1, xmax+1)
axhline(0.0, color='k', alpha=0.1)
legend()
savefig('single_spec_1.pdf')
```

```python
#- Show non-diagonalized covariance matrix
set_cmap('RdBu_r')
subplot(131)  
imshow(N.linalg.inv(iCov), origin='upper', vmin=-3, vmax=3)
xlabel('Covariance')
subplot(132)
imshow(iCov, origin='upper')
xlabel('Inverse Covariance')
subplot(133)
imshow(R, origin='upper')
xlabel('Resolution Matrix')
```

# Example 2: Coadd unaligned spectra

```python
#- Generate two spectra with different x sampling
N.random.seed(0)
def model(x):
    return N.sin(x)

x1 = N.linspace(0, 2*N.pi, 30)
x2 = N.linspace(0, 2*N.pi, 31)
e1 = 0.1*N.ones(len(x1))
e2 = 0.1*N.ones(len(x2))
y1 = model(x1) + N.random.normal(scale=e1)
y2 = model(x2) + N.random.normal(scale=e2)

errorbar(x1, y1, e1, fmt='bs', capsize=0, alpha=0.5)
errorbar(x2, y2, e2, fmt='cs', capsize=0, alpha=0.5)
xlim(-0.1, 2*N.pi+0.1)
ylim(-1.5, 1.5)
```

```python
#- Coadd the spectra
xc = N.linspace(-0.1, 2*N.pi+0.1, 30)
yt = model(xc)

y = N.concatenate([y1, y2])
err = N.concatenate([e1, e2])
W = N.diag(1.0/err**2)
F1 = getF(x1, xc).toarray()
F2 = getF(x2, xc).toarray()
F = N.vstack([F1, F2])

iCov = F.T.dot(W.dot(F))
yc = N.linalg.solve(iCov, F.T.dot(W.dot(y)))
R, ivar = resolution_from_icov(iCov)
ec = N.sqrt(1/ivar)

errorbar(x1, y1-model(x1), e1, fmt='bs', capsize=0, alpha=0.2, label='Spectrum 1')
errorbar(x2, y2-model(x2), e2, fmt='cs', capsize=0, alpha=0.2, label='Spectrum 2')
plot(xc, yc-model(xc), 'rs-', label='Covariant model')
plot(xc, R.dot(yc-model(xc)), 'gs-', label='Diagonalized model')
title('Residuals from input spectrum')
xlim(-0.1, 2*N.pi+0.1)
# ylim(-0.3, 0.6)
legend()
axhline(0.0, color='k', alpha=0.1)
```

```python
Cov = N.linalg.inv(iCov)
subplot(131)  
imshow(Cov, origin='upper')
xlabel('Covariance')
subplot(132)
imshow(iCov, origin='upper')
xlabel('Inverse Covariance')
subplot(133)
imshow(R, origin='upper')
xlabel('Resolution Matrix')
```

# Example 3: Resample covariant spectra

```python
#- Define a rotation matrix that will correlate spectra
def getA(nflux, nspec, c0, c1):
    """
    unnormalized rotation matrix to correlate spectra
    c0 = amount of correlation between neighboring bins of same spectrum
    c1 = amount of correlation between neighboring spectra
    getA(50, 3, -0.5, 0.2) produced interesting amounts of correlation
    """
    n = nflux*nspec
    A = N.zeros((nspec*nflux, nspec*nflux))
    #- Spectra correlated with themselves
    for i in range(nspec):
        for j in range(nflux):
            k = i*nflux+j
            A[k, k] = 1.0
            if j>0:
                A[k-1, k] = A[k,k-1] = c0
                pass
            
    #- Spectra correlated with their neighbors
    for i in range(nspec-1):
        j = i*nflux
        k = (i+1)*nflux
        A[j:j+nflux, k:k+nflux] = A[k:k+nflux, j:j+nflux] = c1*A[j:j+nflux, j:j+nflux]

    return A

nflux = 30
nspec = 3
A = getA(nflux, nspec, -0.5, 0.2)
```

```python
#- Use A to rotate a diagonal covariance matrix
err = N.ones(nflux*nspec)
Cd = N.diag(err)
C = A.dot( Cd.dot(A.T) )
subplot(121); imshow(A); xlabel('A')
subplot(122); imshow(C); xlabel('Covariance')
```

```python
#- Generate uncorrelated spectra and rotate into correlated space
N.random.seed(0)
yd = N.random.normal(scale=err)

#- Add one spike per spectrum
for i in range(nspec):
    j = int((i+1)*nflux/(nspec+1.0) + i*nflux)
    yd[j] += 30.0

#- Rotate into correlated spectra
y = A.dot(yd).reshape(nspec, nflux)
```

```python
#- Hmm.  In this example we aren't resampling; we are just diagonalizing.
#- Make an example that also resamples.
```

```python
iCov = N.linalg.inv(C)
R, ivar = resolution_from_icov(iCov)
yx = R.dot(y.ravel()).reshape(y.shape)
yerr = N.sqrt(1.0/ivar.reshape(y.shape))

#- Covariant spectra
subplot(211)
ylabel('Correlated Spectra')
for i in range(y.shape[0]):
    plot(y[i])
    
#- Diagonalized spectra
subplot(212)
ylabel('Diagonalized Spectra')
for i in range(yx.shape[0]):
    # plot(yx[i])
    errorbar(range(len(yx[i])), yx[i], yerr[i])
    
savefig('multiple_spectra.pdf')
```

```python
subplot(131)  
imshow(Cov, origin='upper')
xlabel('Covariance')
subplot(132)
imshow(iCov, origin='upper')
xlabel('Inverse Covariance')
subplot(133)
imshow(R, origin='upper')
xlabel('Resolution Matrix')
```

```python
#- Study how different correlations produce different R matrices
def getCR(nflux, nspec, c0, c1):
    A = getA(nflux, nspec, c0, c1)
    Cd = N.diag(N.ones(nspec*nflux))
    C = A.dot( Cd.dot(A.T) )
    iCov = N.linalg.inv(C)
    R, ivar = resolution_from_icov(iCov)
    return C, R

nspec = 3
nflux = 10
figure(figsize=(6,8))

C, R = getCR(nflux, nspec, -0.5, 0.2)
subplot(4,3,1); imshow(C)
subplot(4,3,2); imshow(N.linalg.inv(C))
subplot(4,3,3); imshow(R)
C, R = getCR(nflux, nspec, -0.5, 0.0)
subplot(4,3,4); imshow(C)
subplot(4,3,5); imshow(N.linalg.inv(C))
subplot(4,3,6); imshow(R)
C, R = getCR(nflux, nspec, -0.0, 0.2)
subplot(4,3,7); imshow(C)
subplot(4,3,8); imshow(N.linalg.inv(C))
subplot(4,3,9); imshow(R)
C, R = getCR(nflux, nspec, +0.5, 0.0)
subplot(4,3,10); imshow(C)
subplot(4,3,11); imshow(N.linalg.inv(C))
subplot(4,3,12); imshow(R)

subplot(4,3,1); title('Covariance')
subplot(4,3,2); title('Inverse Covariance')
subplot(4,3,3); title('Resolution Matrix')
```
